```python
df = df[df["source_name"] == "mayoclinic"]
display(df[df["name"] == "Desonide Topical"]["source_url"].iloc[0])
df["match"] = df.apply(lambda x : x.source_name in x.source_url, axis=1)
df = df[df["match"] == True]
```

```python
# for i in range(50):
#     print(df["source_name"].iloc[i], df["source_url"].iloc[i])

print(len(df))

print(df[df["source_name"] == "Radiopaque Agent- Diagnostic (Oral Route, Rectal Route, Intravenous Route, Intra-Arterial Route, Intraspinal Route)\n"]["source_url"])

for i in range(40):
    print("i:", i)
    name = df["name"].iloc[i]
    print(name)
    text = df["raw_html"].iloc[i]
    soup = BeautifulSoup(text)
    content = soup.find("div", attrs={"id" : "main-content"})
    ol = content.find("ol", attrs={"class" : "content"})
    items = ol.find_all("li")
    for item in items:
        print(item.text.strip())
```
