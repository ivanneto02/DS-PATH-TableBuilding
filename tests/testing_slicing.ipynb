```python
df = df[df["source_name"] == "medline"]
# display(df[df["name"] == "Desonide Topical"]["source_url"].iloc[0])
df["match"] = df.apply(lambda x : x.source_name in x.source_url, axis=1)
df = df[df["match"] == True]
```

```python
df.head(5)
```

```python
for i in range(10):
    print("i:", i)
    name = df["name"].iloc[i]
    print(name)
    text = df["raw_html"].iloc[i]
    soup = BeautifulSoup(text, "lxml")
    content = soup.find("div", attrs={"id" : "section-brandname-1"})
    li = content.find_all("li")
    for item in li:
        item.find("sup").decompose()
        try:
            item.find("a").decompose()
        except:
            pass
        print(item.text.strip())

    # for item in li:
    #     print(item.text.strip())
    # for item in items:
    #     print(item.text.strip())
```
